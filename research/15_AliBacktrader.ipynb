# Strategy 
Strategy -> newbar -> Next -> Order -> Notify -> newbar

```python
class testSTG(bt.Strategy):
    def __init__(self):
        pass
    def next(self):
        pass

class testclose(bt.Strategy):
    def __init__(self):
        pass 
    def next(self):
        dataclose = self.datas[0].close[0]
        print(dataclose)

class testclosei(bt.Strategy):
    def __init__(self):
        self.dataclose = self.datas[0].close
    def next(self):
        print(self.dataclose[0])

class testLog(bt.Strategy):
    def __init__(self):
        self.dataclose = self.datas[0].close

    def log(self,txt):
        dt = self.datas[0].datetime.datetime()
        print(f'{dt} | {txt}')

    def next(self):
        x = self.dataclose[0]
        self.log(txt=x)
    
```

```python
cerebro=bt.Cerebro()
```

```python
cerebro.adddata(tsla_daily_parsed)
```

```python
cerebro.addstrategy(testLog)
```

```python
cerebro.run()
```

# Momentum Strategy

if close[-2] > close [-1] --> buy 

if close[-2] < close [-1] --> sell

```python
#Define the Momentum Strategy 
class Momentum(bt.Strategy):
    def __init__(self):
        self.dataclose = self.datas[0].close
    def next(self):
        if self.dataclose[-2]>self.dataclose[-1]:
            self.order = self.buy()
            print(f'buy executed @ {self.dataclose[0]}')

        if self.dataclose[-2]<self.dataclose[-1]:
            self.order = self.sell()
            print(f'sell executed @ {self.dataclose[0]}')

        else:
            self.order = self.close()
            print(f'buy executed @ {self.dataclose[0]}')
```

```python
cerebro=bt.Cerebro()
```

```python
cerebro.adddata(tsla_daily_parsed)
```

```python
cerebro.addstrategy(Momentum)
```

```python
cerebro.run()
```

```python
class indicators(bt.Strategy):
    def __init__(self):
        self.sma = bt.indicators.SMA(self.datas[0].close, period = 200)
        self.rsi = bt.indicators.RSI(self.datas[0].close, period = 14)
    
    def next(self):
        print(f'current price {self.datas[0].close[0]} | SMA {self.sma[0]} | RSI {self.rsi[0]}')
```

```python
cerebro=bt.Cerebro()
cerebro.adddata(tsla_daily_parsed)
cerebro.addstrategy(indicators)
```

```python
cerebro.run()

plt.rcParams['figure.figsize'] = [15, 12]
plt.rcParams.update({'font.size': 12})

cerebro.plot(iplot=False)
```
