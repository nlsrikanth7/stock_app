```python
import yfinance as yf
import datetime as dt
import pandas as pd
import os
```

# Download the 1hr data of stocks

```python
# period="50d"
stock_list = {'AMZN', 'VRTX'}
start_date = dt.datetime(2023,10,10)
end_date = dt.datetime.now()

for stock in stock_list:
    try:
        stock_data = yf.download(tickers="AMZN", start= start_date, end= end_date, interval="1h")
        stock_data.to_csv(f'/Users/shrikanthnallapati/Projects/stock_app/Data/{stock}_1h.csv')
        # For saving into json format
        # stock_data.to_json(f'/Users/shrikanthnallapati/Projects/stock_app/Data/{stock}.json')
    except Exception as e:
        print(f'{stock} ==> {e}')
```

```python
data_folder = '/Users/shrikanthnallapati/Projects/stock_app/Data'
AMZN_1h_prices_df = pd.read_csv(os.path.join(data_folder, "AMZN.csv"))
VRTX_1h_prices_df = pd.read_csv(os.path.join(data_folder, "VRTX.csv"))
AMZN_1h_prices_df
```

# Summary Function - .sum(), .mean(), .std(), .last()

```python
selected_day = AMZN_1h_prices_df.loc['2023-12-21']
selected_day['Open'].mean()
```

```python
# Simple mean
AMZN_1h_prices_df['Adj Close'].mean()

# Rolling Mean of a week (24hrs * 7 days)
AMZN_1h_prices_df['Adj Close'].rolling(24*7).mean()
```

```python
# Plotting the graph for rolling mean
AMZN_1h_prices_df['Adj Close'].rolling(24*7).mean().plot(figsize=(16,6))
```

```python
# Plot the graph for rolling std deviation
AMZN_1h_prices_df['Adj Close'].rolling(24*7).std().plot(figsize=(16,6))
```

```python
# Expanding volume
AMZN_1h_prices_df['Volume'].expanding().mean().plot(figsize=(14,7))
```

# Working with Datetime index

```python
# 1D aggregate
AMZN_1h_prices_df.groupby(pd.Grouper(freq='1D')).agg({'Open':'first',
                                                     'Low':'min',
                                                     'High':'max',
                                                     'Adj Close':'last',
                                                     'Volume':'sum'})
```

```python
# 3D aggregate
AMZN_1h_prices_df.groupby(pd.Grouper(freq='3D')).agg({'Open':'first',
                                                     'Low':'min',
                                                     'High':'max',
                                                     'Adj Close':'last',
                                                     'Volume':'sum'})
```
