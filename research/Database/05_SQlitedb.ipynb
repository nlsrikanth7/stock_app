```python
import sqlalchemy
import pandas as pd
import yfinance as yf
# https://www.youtube.com/watch?v=5bUn-D4eL4k
```

```python
wiki = 'http://en.wikipedia.org/wiki/'
tickersSensex = pd.read_html(wiki+'BSE_SENSEX')[1].Symbol.to_list()
tickersDOW = pd.read_html(wiki+'Dow_Jones_Industrial_Average')[1].Symbol.to_list()
```

```python
tickersDOW
```

```python
def getdata(tickers):
    #create a emply list first
    data = []
    for ticker in tickers:
        data.append(yf.download(ticker).reset_index()) # reset_index to get date as column and not as index
    return data
```

```python
US = getdata(tickersDOW)
```

```python
def createengine(name):
    engine = sqlalchemy.create_engine('sqlite:///'+name)
    return engine
```

```python
USengine = createengine('USA')
```

```python
def TOSQL(frames, symbols, engine):
    for frame, symbol in zip(frames, symbols):
        frame.to_sql(symbol, engine, index=False)
    print('Sucessfully imported data')
```

```python
TOSQL(US, tickersDOW, USengine)
```
